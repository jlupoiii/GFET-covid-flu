```python
# showing plots for data from 2/14/25
# this data was DMSO-based functionalization on b2d2 (batch 2 device)
# the sweeping data was collected with 1 source meter, collecting I_DS and I_GS
# wells a and d were functionalized, and wells b and c were blanks (non-functionalized)
# the time series data was done with 0.01X PBS and antigen
# the time series data was collected with 2 channels, one for well a and one for well d
# the fully functionalized wells have dirac curves for varying PBS concs too
    # 0.001X, 0.01X, 0.1X, 1X
```

```python
import numpy as np
import os
import matplotlib.pyplot as plt 
from scipy.optimize import curve_fit
```

```python
# filenames

# filenames for functionalization
filenames_a_func = ['data/data_2-14-25/b2d2a-id.csv', 
                 'data/data_2-14-25/b2d2a-pbase.csv', 
                 'data/data_2-14-25/b2d2a-ab.csv',
                 'data/data_2-14-25/b2d2a-peg.csv',
                 'data/data_2-14-25/b2d2a-1x.csv',]

filenames_b_func = ['data/data_2-14-25/b2d2b-id(blank).csv', 
                 'data/data_2-14-25/b2d2b-pbase(blank).csv', 
                 'data/data_2-14-25/b2d2b-ab(blank).csv',
                 'data/data_2-14-25/b2d2b-peg(blank).csv',
                 'data/data_2-14-25/b2d2b-1x(blank).csv',]

filenames_c_func = ['data/data_2-14-25/b2d2c-id(blank).csv', 
                 'data/data_2-14-25/b2d2c-pbase(blank).csv', 
                 'data/data_2-14-25/b2d2c-ab(blank).csv',
                 'data/data_2-14-25/b2d2c-peg(blank).csv',
                 'data/data_2-14-25/b2d2c-1x(blank).csv',]

filenames_d_func = ['data/data_2-14-25/b2d2d-id.csv', 
                 'data/data_2-14-25/b2d2d-pbase.csv', 
                 'data/data_2-14-25/b2d2d-ab.csv',
                 'data/data_2-14-25/b2d2d-peg.csv',
                 'data/data_2-14-25/b2d2d-1x.csv',]


# filenames for concentrations of PBS
filenames_a_pbs = ['data/data_2-14-25/b2d2a-p001x.csv', 
                 'data/data_2-14-25/b2d2a-p01x.csv', 
                 'data/data_2-14-25/b2d2a-p1x.csv',
                 'data/data_2-14-25/b2d2a-1x.csv',]

filenames_b_pbs = ['data/data_2-14-25/b2d2b-p001x(blank).csv', 
                 'data/data_2-14-25/b2d2b-p01x(blank).csv', 
                 'data/data_2-14-25/b2d2b-p1x(blank).csv',
                 'data/data_2-14-25/b2d2b-1x(blank).csv',]

filenames_c_pbs = ['data/data_2-14-25/b2d2c-p001x(blank).csv', 
                 'data/data_2-14-25/b2d2c-p01x(blank).csv', 
                 'data/data_2-14-25/b2d2c-p1x(blank).csv',
                 'data/data_2-14-25/b2d2c-1x(blank).csv',]

filenames_d_pbs = ['data/data_2-14-25/b2d2d-p001x.csv', 
                 'data/data_2-14-25/b2d2d-p01x.csv', 
                 'data/data_2-14-25/b2d2d-p1x.csv',
                 'data/data_2-14-25/b2d2d-1x.csv',]

filenames_p01X_PBS = ['data/data_2-14-25/b2d2a-p01x.csv',
                     'data/data_2-14-25/b2d2d-p01x.csv']
```

```python
def load_csv_data_1ch(filenames):
    '''
    for multiple file
    6, 4
    '''
    data_list = []
    for filename in filenames:
        data_device = np.genfromtxt(filename, delimiter=',', skip_header=1, filling_values=np.nan).T[[6, 4], :]
        data_device = np.nan_to_num(data_device)
        data_list.append(data_device)
    data = np.array(data_list)
    data[:, 1, :] *= 1000.0
    return data[:,:,-200:][:, :, ::-1] # (num_device, (voltages/currents), voltage_sweep)

def get_dirac_points(IV_mx):
    dirac_idx = np.argmin(IV_mx[:,1,:], axis=-1)
    dirac_ivs = np.array([IV_mx[idx,:,dirac_idx[idx]] for idx in range(len(dirac_idx))])
    return dirac_ivs

data_b2d2a_func = load_csv_data_1ch(filenames_a_func)
data_b2d2a_dirac = get_dirac_points(data_b2d2a_func)

data_b2d2b_func = load_csv_data_1ch(filenames_b_func)
data_b2d2b_dirac = get_dirac_points(data_b2d2b_func)

data_b2d2c_func = load_csv_data_1ch(filenames_c_func)
data_b2d2c_dirac = get_dirac_points(data_b2d2c_func)

data_b2d2d_func = load_csv_data_1ch(filenames_d_func)
data_b2d2d_dirac = get_dirac_points(data_b2d2d_func)
print(data_b2d2d_func.shape)
print(data_b2d2d_dirac.shape)
print('---')


data_b2d2a_pbs = load_csv_data_1ch(filenames_a_pbs)
data_b2d2a_dirac_pbs = get_dirac_points(data_b2d2a_pbs)

data_b2d2b_pbs = load_csv_data_1ch(filenames_b_pbs)
data_b2d2b_dirac_pbs = get_dirac_points(data_b2d2b_pbs)

data_b2d2c_pbs = load_csv_data_1ch(filenames_c_pbs)
data_b2d2c_dirac_pbs = get_dirac_points(data_b2d2c_pbs)

data_b2d2d_pbs = load_csv_data_1ch(filenames_d_pbs)
data_b2d2d_dirac_pbs = get_dirac_points(data_b2d2d_pbs)
print(data_b2d2d_pbs.shape)
print(data_b2d2d_dirac_pbs.shape)
```

```python
def plot_func_steps(dirac_mx_func_list, dirac_mx_func_blank_list, title):

    xlabels = ['Bare graphene', '+PBASE', '+Antibody', '+PEG', '+ETA']
    labels = ['DMSO-functionalized', 'non-functionalized', 'DMF functionalization']
    colors = ['blue', 'red', 'orange']
    for dev_num in range(len(dirac_mx_func_list)):
        plt.scatter(range(dirac_mx_func_list[dev_num][:,0].size), dirac_mx_func_list[dev_num][:,0] - dirac_mx_func_list[dev_num][0,0], s=50, color=colors[0], label=labels[0])
        plt.plot(range(dirac_mx_func_list[dev_num][:,0].size), dirac_mx_func_list[dev_num][:,0] - dirac_mx_func_list[dev_num][0,0], color=colors[0], label=labels[0])

        plt.scatter(range(dirac_mx_func_blank_list[dev_num][:,0].size), dirac_mx_func_blank_list[dev_num][:,0] - dirac_mx_func_blank_list[dev_num][0,0], s=50, color=colors[1], label=labels[1])
        plt.plot(range(dirac_mx_func_blank_list[dev_num][:,0].size), dirac_mx_func_blank_list[dev_num][:,0] - dirac_mx_func_blank_list[dev_num][0,0], color=colors[1], label=labels[1])

    DMF_data = [0, -0.07, -0.055, -0.035, 0.005]
    plt.scatter(range(len(DMF_data)), DMF_data, s=50, color=colors[2], label=labels[2])
    plt.plot(range(len(DMF_data)), DMF_data, color=colors[2], label=labels[2])
    
    handles, labels = plt.gca().get_legend_handles_labels()
    unique = {}
    for handle, label in zip(handles, labels):
        if label not in unique:
            unique[label] = handle
    # sorted_unique = dict(sorted(unique.items(), key=lambda x: x[0]))
    # plt.legend(sorted_unique.values(), sorted_unique.keys())
    plt.legend(unique.values(), unique.keys())

    plt.xlabel('Functionalization Step')
    plt.ylabel(rf'Dirac Shift $(\Delta V)$')
    plt.title(title)
    plt.xticks(range(len(xlabels)), xlabels)
    plt.show()
    plt.close()

plot_func_steps([data_b2d2a_dirac, data_b2d2d_dirac], [data_b2d2b_dirac, data_b2d2c_dirac], 'Dirac Shift for DMSO/DMF Functionalization\n'+rf'$200 \mu m$ channel')
```

```python
def plot_pbs_dirac(dirac_mx_pbs_list, dirac_mx_pbs_blank_list, title):
    xlabels = ['0.001X', '0.01X', '0.1X', '1X']
    colors = ['red', 'blue']
    labels = ['functionalized', 'non-functionalized']
    colors = ['blue', 'red']
    
    for dev_num in range(len(dirac_mx_pbs_list)):
        plt.scatter(range(dirac_mx_pbs_list[dev_num][:,0].size), dirac_mx_pbs_list[dev_num][:,0], s=50, color=colors[0], label=labels[0])
        plt.plot(range(dirac_mx_pbs_list[dev_num][:,0].size), dirac_mx_pbs_list[dev_num][:,0], color=colors[0], label=labels[0])

        plt.scatter(range(dirac_mx_pbs_blank_list[dev_num][:,0].size), dirac_mx_pbs_blank_list[dev_num][:,0], s=50, color=colors[1], label=labels[1])
        plt.plot(range(dirac_mx_pbs_blank_list[dev_num][:,0].size), dirac_mx_pbs_blank_list[dev_num][:,0], color=colors[1], label=labels[1])

    handles, labels = plt.gca().get_legend_handles_labels()
    unique = {}
    for handle, label in zip(handles, labels):
        if label not in unique:
            unique[label] = handle
    sorted_unique = dict(sorted(unique.items(), key=lambda x: x[0]))
    plt.legend(sorted_unique.values(), sorted_unique.keys())

    plt.xlabel('Concentration PBS')
    plt.ylabel('Dirac Point (V)')
    plt.title(title)
    plt.xticks(range(len(xlabels)), xlabels)
    plt.show()
    plt.close()


plot_pbs_dirac([data_b2d2a_dirac_pbs, data_b2d2d_dirac_pbs], [data_b2d2b_dirac_pbs, data_b2d2c_dirac_pbs], 'PBS Concentration vs Dirac Point\nDMSO-based Functionalized Chip\n'+rf'$200 \mu m$ channel')
```

```python
def plot_time_series_1ch(file_path, max_time_sample, xval_action_list, title):
    data_timesweep = np.genfromtxt(file_path, delimiter=",", dtype=np.float64, filling_values=np.nan)
    data_timesweep = data_timesweep[1:max_time_sample,[4,3]].T
    
    data_timesweep[1,:] *= 1000.0
    
    plt.figure(figsize=(10, 4))
    
    minim = np.min(data_timesweep[1])
    maxim = np.max(data_timesweep[1])
    for xval in xval_action_list:
        plt.plot([xval, xval], [minim, maxim], color='r')
    
    plt.plot(data_timesweep[0], data_timesweep[1])
    plt.xlabel('Time (s)')
    plt.ylabel('Drain Current (mA)')
    plt.title(title)
    plt.show()
    plt.close()


file_path = 'data/data_2-11-25/b2d3b-timeSeries-01x.csv'
max_time_sample = 9096
xval_action_list = [128, 340, 411, 500, 604, 699, 800]
title = 'Time-series Drain Current at Fixed Gate Voltage\nGate Voltage=150mV'
plot_time_series_1ch(file_path=file_path, 
                     max_time_sample=max_time_sample, 
                     xval_action_list=xval_action_list, 
                     title=title)
```
