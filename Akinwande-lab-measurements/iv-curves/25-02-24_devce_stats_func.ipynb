```python
# showing plots for data from 2/24/25
# the sweeping data was collected with 1 source meter, collecting I_DS and I_GS
# this data was DMF-based antibody functionalization on b3d1 and b3d3
# the time series data was done only using 1 channel
# the timeseries was done with 10X PBS
```

```python
import numpy as np
import os
import matplotlib.pyplot as plt 
from scipy.optimize import curve_fit
```

```python
def plot_time_series_2ch(file_path, max_time_sample, xval_action_list, title1, title2):
    data_timesweep = np.genfromtxt(file_path, delimiter=",", dtype=np.float64, filling_values=np.nan)

    # first well
    data_timesweep_1 = data_timesweep[1:max_time_sample,[4,3]].T
    data_timesweep_1[1,:] *= 1000.0
    
    plt.figure(figsize=(10, 4))
    
    minim = np.min(data_timesweep_1[1])
    maxim = np.max(data_timesweep_1[1])
    for xval in xval_action_list:
        plt.plot([xval, xval], [minim, maxim], color='r')
    
    plt.plot(data_timesweep_1[0], data_timesweep_1[1])
    plt.xlabel('Time (s)')
    plt.ylabel('Drain Current (mA)')
    plt.title(title1+'\nBare Graphene')
    plt.show()
    plt.close()

    # second well
    data_timesweep_2 = data_timesweep[1:max_time_sample,[4,9]].T
    data_timesweep_2[1,:] *= 1000.0
    
    plt.figure(figsize=(10, 4))
    
    minim = np.min(data_timesweep_2[1])
    maxim = np.max(data_timesweep_2[1])
    for xval in xval_action_list:
        plt.plot([xval, xval], [minim, maxim], color='r')

    plt.plot(data_timesweep_2[0], data_timesweep_2[1])
    plt.xlabel('Time (s)')
    plt.ylabel('Drain Current (mA)')
    plt.title(title2+'\nFunctionalized Graphene')
    plt.show()
    plt.close()




file_path = 'data/data_2-24-25/b3d1-bsa-timeseries.csv'
max_time_sample = 2159
xval_action_list = [20, 65, 120, 180] # GOOD
title1 = 'BSA\nGate Voltage=240mV\n1X PBS'
title2 = 'BSA\nGate Voltage=240mV\n1X PBS'
plot_time_series_2ch(file_path=file_path, 
                     max_time_sample=max_time_sample, 
                     xval_action_list=xval_action_list, 
                     title1=title1,
                     title2=title2)

file_path = 'data/data_2-24-25/b3d1-antigen-timeseries.csv'
max_time_sample = 2954
xval_action_list = [40, 110, 190, 250] # GOOD
title1 = 'Antigen\nGate Voltage=240mV\n1X PBS'
title2 = 'Antigen\nGate Voltage=240mV\n1X PBS'
plot_time_series_2ch(file_path=file_path, 
                     max_time_sample=max_time_sample, 
                     xval_action_list=xval_action_list, 
                     title1=title1,
                     title2=title2)


file_path = 'data/data_2-24-25/b3d3-antigen-timeseries.csv'
max_time_sample = 3655
xval_action_list = [100, 180, 250, 350]
title1 = 'BSA\nGate Voltage=125mV\n10X PBS'
title2 = 'BSA\nGate Voltage=125mV\n10X PBS'
plot_time_series_2ch(file_path=file_path, 
                     max_time_sample=max_time_sample, 
                     xval_action_list=xval_action_list, 
                     title1=title1,
                     title2=title2)

file_path = 'data/data_2-24-25/b3d3-antigen-timeseries.csv'
max_time_sample = 3851
xval_action_list = [100, 180, 250, 350]
title1 = 'Antigen\nGate Voltage=125mV\n10X PBS'
title2 = 'Antigen\nGate Voltage=125mV\n10X PBS'
plot_time_series_2ch(file_path=file_path, 
                     max_time_sample=max_time_sample, 
                     xval_action_list=xval_action_list, 
                     title1=title1,
                     title2=title2)
```
